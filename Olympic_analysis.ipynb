```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
dataset_path = 'Summer-Olympic-medals-1976-to-2008.csv'
```

```python
if not os.path.exists(dataset_path):
    raise FileNotFoundError(f"Dataset not found: {dataset_path}. Please ensure the file is in the correct directory.")
```

```python
df = pd.read_csv(dataset_path, encoding='latin1')
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df.dropna(subset=['Year'], inplace=True)
df['Year'] = df['Year'].astype(int)
```

```python
print(df.info())
print(df.head())
```

```python
print(df.isnull().sum())

df.fillna("Unknown", inplace=True)
```

```python
country_medals = df['Country'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=country_medals.index, y=country_medals.values, palette='viridis')
plt.xlabel("Country")
plt.ylabel("Total Medals")
plt.title("Top 10 Countries by Total Olympic Medals")
plt.xticks(rotation=45)
plt.show()
```

```python
yearly_medals = df.groupby('Year')['Medal'].count()
plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_medals.index.astype(int), y=yearly_medals.values, marker='o', color='b')
plt.xlabel("Year")
plt.ylabel("Total Medals")
plt.title("Olympic Medals Trend Over the Years")
plt.grid(True)
plt.show()
```

```python
medal_counts = df['Medal'].value_counts()
plt.figure(figsize=(7, 5))
sns.barplot(x=medal_counts.index, y=medal_counts.values, palette=['gold', 'silver', 'brown'])
plt.xlabel("Medal Type")
plt.ylabel("Count")
plt.title("Distribution of Medals")
plt.show()

print("Data Analysis Completed!")
```
